```python
# extract genres
import ast
def convert(obj):
    Lis=[]
    for i in ast.literal_eval(obj):
        Lis.append(i['name'])
    return Lis
```

```python
# extract top 3 cast

def convert3(obj):
    Lis=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3 :
            Lis.append(i['name'])
            counter +=1
        else:
            break
        
    return Lis
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
movies.head()
```

```python
# extract director

def fetch_director(obj):
    Lis=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director' :
            Lis.append(i['name'])
            break
    return Lis
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split()) # covert string to list 
```

```python
movies.head(2)
```

```python
# combine the names of same person without space
```

```python
movies['genres']=movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# applying stemming on tags
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem) # stemming
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 ,stop_words= 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
# using cosine distance -- distance between two movies
# distance is inversly propostional to similiarty
# high similiarty low distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0].shape
```

```python
similarity[1]  # similairty with movie 2 is 1
```

```python
new_df[new_df['title'] == 'Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True , key = lambda x : x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True , key = lambda x : x[1])[1:6]
    
    for i in movies_list :
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
